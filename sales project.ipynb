```python
import pandas as pd
coustmer=({"coustmerid":[12,14,44],"name":["shreyanshi","isha","ash"]})
order=({"orderid":[1201,1401,4402],"coustmerid":[12,14,44],"amount":[25000,5000,6000]})
dfc=pd.DataFrame(coustmer)
dfo=pd.DataFrame(order)
dfm=pd.merge(dfc,dfo,on='coustmerid')
print(dfc)
print(dfo)
print(dfm)
result=dfm.groupby('orderid')['amount'].sum()
dfr=pd.DataFrame(result)
print(dfr)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
y=[9,11,23,34,45,56,33,24,35,56,55,34]
y1=[3,14,12,34,12,34,45,55,56,34,23,23]
plt.plot(x,y,color="blue",linestyle='--',marker='o')
plt.plot(x,y1,color='red',linestyle=':',marker='^')
plt.xlabel("months")
plt.ylabel("sales per month by two shops")
plt.title("LINE PLOT TREND COMPARISION BETWEEN TWO SHOPS")
plt.show()
```
